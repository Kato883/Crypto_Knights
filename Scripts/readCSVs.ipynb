```python
import pandas as pd
from datetime import datetime
```

```python
bitcoin_file = "CSVs/bitcoin_price.csv"
ethereum_file = "CSVs/ethereum_price.csv"
litecoin_file = "CSVs/litecoin_price.csv"
ripple_file = "CSVs/ripple_price.csv"
dash_file = "CSVs/dash_price.csv"
```

```python
bitcoin_file_df = pd.read_csv(bitcoin_file, encoding="ISO-8859-1")
ethereum_file_df = pd.read_csv(ethereum_file, encoding="ISO-8859-1")
litecoin_file_df = pd.read_csv(litecoin_file, encoding="ISO-8859-1")
ripple_file_df = pd.read_csv(ripple_file, encoding="ISO-8859-1")
dash_file_df = pd.read_csv(dash_file, encoding="ISO-8859-1")
```

```python
bitcoin_file_df['Symbol'] = 'BTC'
new_bitcoin_file_df = bitcoin_file_df
new_bitcoin_file_df.head()
```

```python
new_bitcoin_file_df['Volume'] = new_bitcoin_file_df['Volume'].str.replace(",", "")
new_bitcoin_file_df['Volume'] = new_bitcoin_file_df['Volume'].str.replace("-", "0").astype('int64')
new_bitcoin_file_df['Market Cap'] = new_bitcoin_file_df['Market Cap'].str.replace(",", "")
new_bitcoin_file_df['Market Cap'] = new_bitcoin_file_df['Market Cap'].str.replace("-", "0").astype('int64')
new_bitcoin_file_df['Close Increase'] = new_bitcoin_file_df['Close'] - new_bitcoin_file_df['Close'].shift(-1)
new_bitcoin_file_df['Volume Increase'] = new_bitcoin_file_df['Volume'] - new_bitcoin_file_df['Volume'].shift(-1)
new_bitcoin_file_df['Market Cap Increase'] = new_bitcoin_file_df['Market Cap'] - new_bitcoin_file_df['Market Cap'].shift(-1)
final_bitcoin_df = new_bitcoin_file_df[['Date', 'Symbol', 'Open', 'High', 'Low', 'Close', 'Close Increase', 'Volume', 'Volume Increase', 'Market Cap', 'Market Cap Increase']]
final_bitcoin_df.head(10)
```

```python
ethereum_file_df['Symbol'] = 'ETH'
new_ethereum_file_df = ethereum_file_df
new_ethereum_file_df.head(10)
```

```python
new_ethereum_file_df['Volume'] = new_ethereum_file_df['Volume'].str.replace(",", "").astype('int64')
new_ethereum_file_df['Market Cap'] = new_ethereum_file_df['Market Cap'].str.replace(",", "")
new_ethereum_file_df['Market Cap'] = new_ethereum_file_df['Market Cap'].str.replace("-", "0").astype('int64')
new_ethereum_file_df['Close Increase'] = new_ethereum_file_df['Close'] - new_ethereum_file_df['Close'].shift(-1)
new_ethereum_file_df['Volume Increase'] = new_ethereum_file_df['Volume'] - new_ethereum_file_df['Volume'].shift(-1)
new_ethereum_file_df['Market Cap Increase'] = new_ethereum_file_df['Market Cap'] - new_ethereum_file_df['Market Cap'].shift(-1)
final_ethereum_df = new_ethereum_file_df[['Date', 'Symbol', 'Open', 'High', 'Low', 'Close', 'Close Increase', 'Volume', 'Volume Increase', 'Market Cap', 'Market Cap Increase']]
final_ethereum_df.head(10)
```

```python
new_litecoin_file_df['Volume'] = new_litecoin_file_df['Volume'].str.replace(",", "")
new_litecoin_file_df['Volume'] = new_litecoin_file_df['Volume'].str.replace("-", "0").astype('int64')
new_litecoin_file_df['Market Cap'] = new_litecoin_file_df['Market Cap'].str.replace(",", "").astype('int64')
new_litecoin_file_df['Close Increase'] = new_litecoin_file_df['Close'] - new_litecoin_file_df['Close'].shift(-1)
new_litecoin_file_df['Volume Increase'] = new_litecoin_file_df['Volume'] - new_litecoin_file_df['Volume'].shift(-1)
new_litecoin_file_df['Market Cap Increase'] = new_litecoin_file_df['Market Cap'] - new_litecoin_file_df['Market Cap'].shift(-1)
final_litecoin_df = new_litecoin_file_df[['Date', 'Symbol', 'Open', 'High', 'Low', 'Close', 'Close Increase', 'Volume', 'Volume Increase', 'Market Cap', 'Market Cap Increase']]
final_litecoin_df.head(10)
```

```python
ripple_file_df['Symbol'] = 'XRP-USD'
new_ripple_file_df = ripple_file_df
new_ripple_file_df.head(10)
```

```python
new_ripple_file_df['Volume'] = new_ripple_file_df['Volume'].str.replace(",", "")
new_ripple_file_df['Volume'] = new_ripple_file_df['Volume'].str.replace("-", "0").astype('int64')
new_ripple_file_df['Market Cap'] = new_ripple_file_df['Market Cap'].str.replace(",", "").astype('int64')
new_ripple_file_df['Close Increase'] = new_ripple_file_df['Close'] - new_ripple_file_df['Close'].shift(-1)
new_ripple_file_df['Volume Increase'] = new_ripple_file_df['Volume'] - new_ripple_file_df['Volume'].shift(-1)
new_ripple_file_df['Market Cap Increase'] = new_ripple_file_df['Market Cap'] - new_ripple_file_df['Market Cap'].shift(-1)
final_ripple_df = new_ripple_file_df[['Date', 'Symbol', 'Open', 'High', 'Low', 'Close', 'Close Increase', 'Volume', 'Volume Increase', 'Market Cap', 'Market Cap Increase']]
final_ripple_df.head(10)
```

```python
dash_file_df['Symbol'] = 'DASH-USD'
new_dash_file_df = dash_file_df
new_dash_file_df.head(10)
```

```python
new_dash_file_df['Volume'] = new_dash_file_df['Volume'].str.replace(",", "").astype('int64')
new_dash_file_df['Market Cap'] = new_dash_file_df['Market Cap'].str.replace(",", "").astype('int64')
new_dash_file_df['Close Increase'] = new_dash_file_df['Close'] - new_dash_file_df['Close'].shift(-1)
new_dash_file_df['Volume Increase'] = new_dash_file_df['Volume'] - new_dash_file_df['Volume'].shift(-1)
new_dash_file_df['Market Cap Increase'] = new_dash_file_df['Market Cap'] - new_dash_file_df['Market Cap'].shift(-1)
final_dash_df = new_dash_file_df[['Date', 'Symbol', 'Open', 'High', 'Low', 'Close', 'Close Increase', 'Volume', 'Volume Increase', 'Market Cap', 'Market Cap Increase']]
final_dash_df.head(10)
```

```python
crypto_frames = [final_bitcoin_df, final_ethereum_df, final_litecoin_df, final_ripple_df, final_dash_df]
```

```python
crypto_currency_df = pd.concat(crypto_frames)
crypto_currency_df.fillna(0, inplace=True)
```

```python
crypto_currency_df['Date'] = pd.to_datetime(crypto_currency_df['Date'], dayfirst = False, yearfirst = False)
crypto_currency_df['Date'] = pd.to_datetime(crypto_currency_df['Date'], format='%m-%d-%Y')
crypto_currency_df['Date2'] = crypto_currency_df['Date'].dt.strftime('%m-%d-%Y')
final_crypto_currency_df = crypto_currency_df[['Date2', 'Symbol', 'Open', 'High', 'Low', 'Close', 'Close Increase', 'Volume', 'Volume Increase', 'Market Cap', 'Market Cap Increase']]
final_crypto_currency_df.rename(columns={'Date2': 'Date'}, inplace=True)

final_crypto_currency_df
```
